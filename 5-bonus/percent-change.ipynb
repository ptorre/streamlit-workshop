# Creating a "Difference" Column

You can create a new column by simply assigning to it. This new column can be the result of computations between other columns:

```python
df_wide = be.get_data_wide("Total Population")

cols = ["State", "State Abbrev", "2005", "2023"]
df_wide = df_wide[cols]

df_wide["Diff"] = df_wide["2023"] - df_wide["2005"]
df_wide.head()
```

# Maps

Below is code to map the percent change in Population between 2005 and 2023:

```python
import plotly.express as px

df_wide = be.get_data_wide("Total Population")

cols = ["State", "State Abbrev", "2005", "2023"]
df_wide = df_wide[cols]

df_wide["Diff"] = (df_wide["2023"] - df_wide["2005"]) / df_wide["2005"] * 100

px.choropleth(
    df_wide,
    locations="State Abbrev",  # Column for region
    locationmode="USA-states",
    color="Diff",  # Column for color
    scope="usa",
    color_continuous_scale="Viridis",
    title="Percent Change in Total Population: 2005-2023",
)
```
